## Detect and draw inner circles

```python
for img in images:
    circle = hough.detect_inner_circle(img)
    cv2.circle(img,(circle[0],circle[1]),circle[2],(0,255,0),2)
    cv2.circle(img,(circle[0],circle[1]),2,(0,255,0),3)
```
